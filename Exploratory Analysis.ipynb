# **Analysed the most important features of a diamonds dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv('./data/diamonds_train.csv')
```

```python
diamonds.describe()
```

```python
diamonds.info()
```

# **Exploratory Analysis of Numeric Data**

```python
diamonds.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

### **Price by Size and Color**

It is observed that the D color are more expensive with smaller diamond size while H and I need a larger diamonf to reach the same price
D color have a exponential price increase with size while I and H have logarithmic

```python
fig, ax = plt.subplots(nrows = 3, ncols = 1, figsize = (20, 20))

sns.scatterplot(x = 'x', y = 'price', hue = 'color', data = diamonds, ax = ax[0])
sns.scatterplot(x = 'y', y = 'price', hue = 'color', data = diamonds, ax = ax[1])
ax[1].set_xlim([0, 15])
sns.scatterplot(x = 'z', y = 'price', hue = 'color', data = diamonds, ax = ax[2])

plt.savefig('./images/size_price.png')
```

### Price by carat

It is observed that the diamonds with clarity IF are the most expensive with lower carat.

It is also observed that diamonds of color D are the most expensive at lower carat

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (20,15))

sns.scatterplot(x = 'carat', y = 'price', hue = 'clarity', data = diamonds, ax = ax[0])
sns.scatterplot(x = 'carat', y = 'price', hue = 'color', data = diamonds, ax = ax[1])

plt.savefig('./images/carat_price.png')
```

# **Exploratory Analysis of Categorical Data**


### **How Color Affect Price**

The most expensives Diamonds are J color followed by I, H and G. 
Taking into account the cut type, the most expensive cut is Premium and the worst is Ideal

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (25, 10))


sns.barplot(x = 'color', y = 'price', data = diamonds, ax = ax[0])
sns.barplot(x = 'color', y = 'price', data = diamonds, hue = 'cut', ax = ax[1])

plt.savefig('./images/color_price.png')
```

### **How Clarity Affect Price**

The most expensives Diamonds are SI2 clarity, followed by VS2, SI1 and I1.
Taking into account color and clarity the most expensive diamond is a diamond of color D and clarity IF.

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (30, 10))

sns.barplot(x = 'clarity', y = 'price', data = diamonds, ax = ax[0])
sns.barplot(x = 'clarity', y = 'price', data = diamonds, hue = 'color', ax = ax[1])

plt.savefig('./images/clarity_price.png')
```

# **Conclusions**

The diamonds of color D and clarity IF are the most expensive. Their prices grow exponentially with carat and size. 

However, this diamonds are also the most strange ones because when we compare the mean prices their mean price is one of the lowest. Therefore, this kind of diamonds are really expensive but it is extremelly weird to find huge and clear diamonds of this type.

The cheapest diamonds are diamonds with SI2 clarity and J color. Their price grow linearly. However, when we check the mean price of these diamonds it is observed that they are the highest one, thus is due to the fact that this diamonds are commonly found in huge shapes and clearness.
